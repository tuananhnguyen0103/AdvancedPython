```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
image_path = "hot_girl.jpg"  # Thay i thnh ng dn thc t ca hnh nh
image_np = plt.imread(image_path)
plt.imshow(image_np)
# plt.axis('off')  # Turn off axes
plt.show()
```

```python
print(image_np)
print(image_np.ndim)
print(image_np.size)
print(image_np.shape)
```

```python


# Chuyn nh mu sang nh xm v nh en trng
gray_image = np.mean(image_np, axis=2)
threshold = 128
black_white_image = np.where(gray_image > threshold, 255, 0)
image_r = image_np[:,:,0]
image_g =image_np[:,:,1]
image_b = image_np[:,:,2]
# To mt bn khu vc hin th (1 hng, 3 ct)
plt.subplot(1, 3, 1)
plt.imshow(image_r)
plt.title("nh Mu")
# plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image_g, cmap='gray')
plt.title("nh Mu 2")
# plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(image_b, cmap='gray', vmin=0, vmax=255)
plt.title("nh Mu 3")
# plt.axis('off')

# Hin th tt c bc nh
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# To d liu mu
x = np.linspace(0, 10, 100)
y = np.sin(x)

# To  th
plt.plot(x, y)
plt.xlabel("Trc X")
plt.ylabel("Trc Y")
plt.title(" th hm sin(x)")

# Ty chnh nhn trc x v y
x_ticks = np.arange(0, 11, 2)  # Danh sch cc s trn trc x
y_ticks = np.linspace(-1, 1, 5)  # Danh sch cc s trn trc y
plt.xticks(x_ticks)
plt.yticks(y_ticks)

# Hin th  th
plt.show()
```
