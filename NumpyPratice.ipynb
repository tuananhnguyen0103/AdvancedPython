```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
image_path = "hot_girl.jpg"  # Thay đổi thành đường dẫn thực tế của hình ảnh
image_np = plt.imread(image_path)
plt.imshow(image_np)
# plt.axis('off')  # Turn off axes
plt.show()
```

```python
print(image_np)
print(image_np.ndim)
print(image_np.size)
print(image_np.shape)
```

```python


# Chuyển ảnh màu sang ảnh xám và ảnh đen trắng
gray_image = np.mean(image_np, axis=2)
threshold = 128
black_white_image = np.where(gray_image > threshold, 255, 0)
image_r = image_np[:,:,0]
image_g =image_np[:,:,1]
image_b = image_np[:,:,2]
# Tạo một bốn khu vực hiển thị (1 hàng, 3 cột)
plt.subplot(1, 3, 1)
plt.imshow(image_r)
plt.title("Ảnh Màu")
# plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image_g, cmap='gray')
plt.title("Ảnh Màu 2")
# plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(image_b, cmap='gray', vmin=0, vmax=255)
plt.title("Ảnh Màu 3")
# plt.axis('off')

# Hiển thị tất cả bức ảnh
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo dữ liệu mẫu
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Tạo đồ thị
plt.plot(x, y)
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.title("Đồ thị hàm sin(x)")

# Tùy chỉnh nhãn trục x và y
x_ticks = np.arange(0, 11, 2)  # Danh sách các số trên trục x
y_ticks = np.linspace(-1, 1, 5)  # Danh sách các số trên trục y
plt.xticks(x_ticks)
plt.yticks(y_ticks)

# Hiển thị đồ thị
plt.show()
```
