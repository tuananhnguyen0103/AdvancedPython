# Bài 2

```python
import pandas as pd
import numpy as np

# Dữ liệu ban đầu
s = pd.Series([8.5, 9.0, 7.0, 3.0, 'CĐG', 6.0, 7.5, 'CĐG', 8.0])

# a) Chuyển kiểu dữ liệu trong s thành dạng số chứa trong mảng a
a = pd.to_numeric(s, errors='coerce')
# In kết quả
print("Dữ liệu a sau khi xử lý:")
print(a)

# b) Thay thế các giá trị không phải số trong a bằng giá trị 0
a = a.fillna(0)
print(a)

# c) Tìm giá trị lớn nhất và nhỏ nhất của a
max_value = a.max()
min_value = a.min()

# d) Tính tổng, giá trị trung bình (mean), giá trị trung vị (median) của a
total = a.sum()
mean = a.mean()
median = a.median()



print("Giá trị lớn nhất của a:", max_value)
print("Giá trị nhỏ nhất của a:", min_value)

print("Tổng của a:", total)
print("Giá trị trung bình của a:", mean)
print("Giá trị trung vị của a:", median)
```

# Bài 3

```python
import pandas as pd
import numpy as np

# a) Tạo một Pandas Series d gồm 20 giá trị random từ 0 đến 1
d = pd.Series(np.random.rand(20))

# b) Sắp xếp d theo giá trị tăng dần
d_sorted_ascending = d.sort_values()

# c) Sắp xếp d theo giá trị giảm dần
d_sorted_descending = d.sort_values(ascending=False)

# d) Trích chọn từ d các phần tử lớn hơn hoặc bằng 0.8
d_greater_than_08 = d[d >= 0.8]

# e) Trích chọn từ d các phần tử nằm trong đoạn [0.5, 0.79]
d_between_05_and_079 = d[(d >= 0.5) & (d <= 0.79)]

# In kết quả
print("a) Series d:")
print(d)


```

```python
print("\nb) Sắp xếp d theo giá trị tăng dần:")
print(d_sorted_ascending)
```

```python
print("\nc) Sắp xếp d theo giá trị giảm dần:")
print(d_sorted_descending)
```

```python
print("\nd) Các phần tử lớn hơn hoặc bằng 0.8:")
print(d_greater_than_08)
```

```python
print("\ne) Các phần tử nằm trong đoạn [0.5, 0.79]:")
print(d_between_05_and_079)
```

# Bài 4

```python
import pandas as pd

# a) Tạo một Pandas DataFrame d để lưu trữ dữ liệu hóa đơn
data = {
    'OrderID': [101, 101, 102, 102, 103],
    'ProductID': ['LAC', 'MLT', 'LAC', 'HPA', 'MMS'],
    'ProductName': ['Acer Laptop', 'Logitech Mouse', 'Acer Laptop', 'Panasonic Headphone', 'Microsoft Mouse'],
    'Quantity': [1, 2, 3, 1, 1],
    'Price': [18900000, 300000, 18900000, 1000000, 250000],
    'Discount': [0.5, 0.3, 0.5, 0.2, 0.6]
}

df = pd.DataFrame(data)

# b) Hiển thị dữ liệu toàn bộ DataFrame ra màn hình
print("Dữ liệu toàn bộ DataFrame:")
print(df)




```

```python
# c) Hiển thị dữ liệu các trường “OrderID, ProductID, ProductName” ra màn hình
print("\nDữ liệu các trường OrderID, ProductID, ProductName:")
print(df[['OrderID', 'ProductID', 'ProductName']])

```

```python
# d) Truy xuất lấy về dữ liệu của cột ProductID
product_ids = df['ProductID']
print("\nDữ liệu của cột ProductID:")
print(product_ids)

```

```python
# e) Truy xuất lấy về dữ liệu của một mã hóa đơn cụ thể
order_101 = df[df['OrderID'] == 101]
print("\nDữ liệu của hóa đơn có OrderID = 101:")
print(order_101)
```

```python

# f) Truy xuất lấy về các mặt hàng đã bán với chiết khấu >=0.5 (50%)
items_with_discount_50 = df[df['Discount'] >= 0.5]
print("\nCác mặt hàng đã bán với chiết khấu >= 0.5:")
print(items_with_discount_50)

```

```python
# g) Tính tổng doanh thu trên toàn bộ sản phẩm bán được
total_revenue = (df['Price'] * df['Quantity'] * (1 - df['Discount'])).sum()
print("\nTổng doanh thu trên toàn bộ sản phẩm bán được:", total_revenue)
```

```python
# h) Tính tổng doanh thu trên mỗi sản phẩm
product_revenue = df.groupby(['ProductID', 'ProductName','Discount','Price'])['Quantity'].sum().reset_index()
print(product_revenue)
product_revenue['Revenue'] = product_revenue['Quantity'] * (1 - product_revenue['Discount']) *product_revenue['Price']  
print("\nTổng doanh thu trên mỗi sản phẩm:")
print(product_revenue[['ProductID', 'ProductName', 'Revenue']])
```

```python
import pandas as pd

# a) Đọc toàn bộ dữ liệu từ tệp styles.csv ra DataFrame và hiển thị ra màn hình
df = pd.read_csv('./data/styles.csv')
print("Dữ liệu toàn bộ từ tệp styles.csv:")
print(df.shape)
df


```

```python
# b) Hiển thị ra các sản phẩm của năm 2019
products_2019 = df[df['year'] == 2019].reset_index()
print(products_2019.shape)
print("\nCác sản phẩm của năm 2019:")
products_2019

#
```

```python
#c) Hiển thị ra các sản phẩm mùa “Winter” năm 2019
winter_2019 = df[(df['season'] == 'Winter') & (df['year'] == 2019)]
print("\nCác sản phẩm mùa Winter năm 2019:")
winter_2019
```

```python

# d) Hiển thị ra các sản phẩm áo “Tshirts” của “Men” năm 2019
tshirts_men_2019 = df[(df['articleType'] == 'Tshirts') & (df['gender'] == 'Men') & (df['year'] == 2019)]
print("\nCác sản phẩm áo Tshirts của Men năm 2019:")
tshirts_men_2019

```

```python
# e) Thống kê số lượng sản phẩm
# (1) Theo giới tính
gender_counts = df['gender'].value_counts()
print("\nThống kê số lượng sản phẩm theo giới tính:")
print(gender_counts)



```

```python
# (2) Theo năm
year_counts = df['year'].value_counts()
print("\nThống kê số lượng sản phẩm theo năm:")
print(year_counts)

```

```python
# (3) Theo loại sản phẩm (articleType)
article_type_counts = df['articleType'].value_counts()
print("\nThống kê số lượng sản phẩm theo loại sản phẩm:")
print(article_type_counts)
```

```python
# f) Thống kê số lượng sản phẩm mỗi mùa của năm 2019 với 2 cột: Season, Quantity
season_2019_counts = products_2019['season'].value_counts().reset_index()
season_2019_counts.columns = ['Season', 'Quantity']
print("\nThống kê số lượng sản phẩm mỗi mùa của năm 2019:")
print(season_2019_counts)
```
