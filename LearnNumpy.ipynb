```python
from PIL import Image
import numpy as np

# Đọc hình ảnh bằng Pillow
image_path = "hot_girl.jpg"  # Thay đổi thành đường dẫn thực tế của hình ảnh
image_pil = Image.open(image_path)

# Chuyển hình ảnh thành mảng NumPy
image_np = np.array(image_pil)

# Hiển thị thông tin về mảng hình ảnh
print("Shape of image array:", image_np.shape)  # Kích thước của mảng hình ảnh
print("Data type of image array:", image_np.dtype)  # Kiểu dữ liệu của mảng hình ảnh

# Hiển thị hình ảnh (sử dụng Matplotlib)
import matplotlib.pyplot as plt
import matplotlib.image as img
plt.imshow(image_np)
plt.axis('off')  # Tắt trục x, y
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Read the image as a NumPy array
image_path = "hot_girl.jpg"  # Replace with the actual image path
image_np_gpt = plt.imread(image_path)

# Define the crop box (y1:y2, x1:x2)
crop_box = (0,600,0,600)  # Example crop box coordinates

# Crop the image using NumPy slicing
cropped_image = image_np_gpt[crop_box[0]:crop_box[1], crop_box[2]:crop_box[3]]

# Display the cropped image
plt.imshow(cropped_image)
plt.axis('off')
plt.show()
```

```python
# r_image = image_np[0]
# g_image = image_np[1]
# b_image = image_np[2]
# Tạo mảng dữ liệu ảnh xám
gray_image = 0.2989 * image_np[:,:,0] + 0.5870 * image_np[:,:,1] + 0.1140 * image_np[:,:,2]

# Hiển thị ảnh xám
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Gán giá trị trắng và đen
black_white_image = gray_image
black_white_image = np.where(black_white_image > 128, 255, 0)
# Hiển thị ảnh đen trắng
plt.imshow(black_white_image, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```
