```python
import numpy as np
import matplotlib.pyplot as plt

# To d liu ngu nhin theo phn phi chun
mean = 6  # Gi tr trung bnh
std_dev = 0.5  #  lch chun
num_samples = 1000  # S mu

data = np.random.normal(mean, std_dev, num_samples)
data = np.clip(data, 0, 10)
# V biu  histogram
plt.hist(data, bins=20, density=True, alpha=0.7, color='b', edgecolor='k')
plt.xlabel('Gi tr')
plt.ylabel('Tn sut')
plt.title('Biu  Histogram ca D liu Ngu nhin')
plt.grid(True)

# Hin th biu 
plt.show()
```

```python
import numpy as np
from scipy.stats import truncnorm

# Define the range and parameters of the truncated normal distribution
min_val = 6
max_val = 10
mean = 0
std_dev = 8.5

# Calculate the lower and upper bounds for the truncation
a = (min_val - mean) / std_dev
b = (max_val - mean) / std_dev

# Generate random data within the specified range
size = 100  # Adjust the number of data points as needed
random_data = truncnorm.rvs(a, b, loc=mean, scale=std_dev, size=size)
plt.hist(random_data, bins=20, density=True, alpha=0.7, color='b', edgecolor='k')
plt.xlabel('Gi tr')
plt.ylabel('Tn sut')
plt.title('Biu  Histogram ca D liu Ngu nhin')
plt.grid(True)

# Hin th biu 
plt.show()
# Now, random_data contains random values within the range [min_val, max_val]
```

```python
import matplotlib.pyplot as plt

# To d liu mu
x = [1, 2, 3, 4]
y1 = [10, 11, 12, 13]
y2 = [20, 21, 22, 23]
y3 = [30, 31, 32, 33]
y4 = [40, 41, 42, 43]

# To subplot 2x2
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Plot 4')

# Hin th  th
plt.tight_layout()  # m bo cc  th khng trn ln nhau
plt.show()
```
