```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo dữ liệu ngẫu nhiên theo phân phối chuẩn
mean = 6  # Giá trị trung bình
std_dev = 0.5  # Độ lệch chuẩn
num_samples = 1000  # Số mẫu

data = np.random.normal(mean, std_dev, num_samples)
data = np.clip(data, 0, 10)
# Vẽ biểu đồ histogram
plt.hist(data, bins=20, density=True, alpha=0.7, color='b', edgecolor='k')
plt.xlabel('Giá trị')
plt.ylabel('Tần suất')
plt.title('Biểu đồ Histogram của Dữ liệu Ngẫu nhiên')
plt.grid(True)

# Hiển thị biểu đồ
plt.show()
```

```python
import numpy as np
from scipy.stats import truncnorm

# Define the range and parameters of the truncated normal distribution
min_val = 6
max_val = 10
mean = 0
std_dev = 8.5

# Calculate the lower and upper bounds for the truncation
a = (min_val - mean) / std_dev
b = (max_val - mean) / std_dev

# Generate random data within the specified range
size = 100  # Adjust the number of data points as needed
random_data = truncnorm.rvs(a, b, loc=mean, scale=std_dev, size=size)
plt.hist(random_data, bins=20, density=True, alpha=0.7, color='b', edgecolor='k')
plt.xlabel('Giá trị')
plt.ylabel('Tần suất')
plt.title('Biểu đồ Histogram của Dữ liệu Ngẫu nhiên')
plt.grid(True)

# Hiển thị biểu đồ
plt.show()
# Now, random_data contains random values within the range [min_val, max_val]
```

```python
import matplotlib.pyplot as plt

# Tạo dữ liệu mẫu
x = [1, 2, 3, 4]
y1 = [10, 11, 12, 13]
y2 = [20, 21, 22, 23]
y3 = [30, 31, 32, 33]
y4 = [40, 41, 42, 43]

# Tạo subplot 2x2
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Plot 4')

# Hiển thị đồ thị
plt.tight_layout()  # Đảm bảo các đồ thị không tràn lên nhau
plt.show()
```
